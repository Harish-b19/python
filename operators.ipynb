```python
# 1.Write a program to perform addition, subtraction, multiplication, division, floor division, modulus, and exponentiation on two numbers.

print("1.Add: ")
print("2.sub: ")
print("3.mul: ")
print("4.div: ")
print("5.floor div: ")
print("6.mod: ")
print("7.expo: ")

n=int(input("choose the above one: "))
x=int(input("enter the x value"))
y=int(input("enter the y value"))

if n==1:
  print("add: ",x+y)
elif n==2:
  print("sub: ",x-y)
elif n==3:
  print("mul: ",x*y)
elif n==4:
  print("div: ",x/y)
elif n==5:
  print("floor div: ",x//y)
elif n==6:
  print("mod: ",x%y)
elif n==7:
  print("expo: ",x**y)
else:
  print("invalid")
```

```python
# 30.Build a menu-driven calculator that uses arithmetic, comparison, logical, assignment, membership, identity, and bitwise operators wherever applicable

while True:
    print("\n===== MENU DRIVEN CALCULATOR =====")
    print("1. Arithmetic Operators")
    print("2. Comparison Operators")
    print("3. Logical Operators")
    print("4. Assignment Operators")
    print("5. Membership Operators")
    print("6. Identity Operators")
    print("7. Bitwise Operators")
    print("8. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))

        print("Addition =", a + b)
        print("Subtraction =", a - b)
        print("Multiplication =", a * b)
        print("Division =", a / b)
        print("Floor Division =", a // b)
        print("Modulus =", a % b)
        print("Power =", a ** b)

    elif choice == 2:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))

        print("a == b :", a == b)
        print("a != b :", a != b)
        print("a > b  :", a > b)
        print("a < b  :", a < b)
        print("a >= b :", a >= b)
        print("a <= b :", a <= b)

    elif choice == 3:
        x = bool(int(input("Enter first boolean (0 or 1): ")))
        y = bool(int(input("Enter second boolean (0 or 1): ")))

        print("x and y =", x and y)
        print("x or y  =", x or y)
        print("not x   =", not x)

    elif choice == 4:
        a = int(input("Enter a number: "))

        print("Original value =", a)

        a += 5
        print("After +=5 :", a)

        a -= 2
        print("After -=2 :", a)

        a *= 3
        print("After *=3 :", a)

        a //= 2
        print("After //=2 :", a)

    elif choice == 5:
        text = input("Enter a string: ")
        ch = input("Enter a character: ")

        print(ch, "in string :", ch in text)
        print(ch, "not in string :", ch not in text)

    elif choice == 6:
        list1 = [10, 20, 30]
        list2 = list1
        list3 = [10, 20, 30]

        print("list1 is list2 :", list1 is list2)
        print("list1 is list3 :", list1 is list3)
        print("list1 == list3 :", list1 == list3)

    elif choice == 7:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))

        print("a & b =", a & b)
        print("a | b =", a | b)
        print("a ^ b =", a ^ b)
        print("~a =", ~a)
        print("a << 1 =", a << 1)
        print("a >> 1 =", a >> 1)

    elif choice == 8:
        print("Calculator Closed.")
        break

    else:
        print("Invalid Choice")
```
